```python
import pandas as pd
df=pd.read_csv('Data.csv')
df = df.set_index('Unnamed: 0')
df
```

```python
df['Rate'].fillna(6)
```

```python
import pandas as pd
import numpy as np
import scipy.stats as sts
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize = (10,5))
df.Rate.hist(bins=50, density=True);
df.Rate.plot(kind='kde', linewidth=4)
plt.xlim(2,10)
plt.title("Распределение оценок");
meanmark=df.Rate.mean()
modemark=df.Rate.mode()
print('Cредняя оценка;',meanmark)
print('Модальная оценка;',modemark)
```

Большинство фильмов имеют рейтинг от 6 до 8. Можно сказать, что фильм с рейтингом больше 8 на Кинопоиске может считаться очень хорошим. 

```python
plt.figure(figsize = (10,5))
df.Duration.hist(bins=50, density=True);
df.Duration.plot(kind='kde', linewidth=4)
plt.xlim(40,250)
plt.title("Распределение продолжительности фильма");
meandur=df.Duration.mean()
modedur=df.Duration.mode()
print('Cредняя продолжительность;',meandur)
print('Мода продолжительности', modedur)
```

В основном люди смотрят полуторачасовые фильмы, но главные выводы о зависимости длительности фильма и его успеха будут похже. 

```python
plt.figure(figsize = (10,5))
df.Year.hist(bins=50, density=True);
df.Year.plot(kind='kde', linewidth=4)
plt.xlim(1960,2023)
plt.title("Распредление фильмов по новизне");
```

Современные люди смотрят современные фильмы. Да и в 21 веке фильмов выходит куда больше.

```python
plt.xlabel('Год выпуска')
plt.ylabel('Рейтинг фильма')
df['Year']=df['Year'].fillna(0)
df['Year']=df['Year'].astype(int)
df['Year']=np.where(df['Year']<1895,2024,df['Year'])
plt.scatter(df['Year'],df['Rate'])
```

Старые фильмы не оцегиваются ниже 6.5. Причина проста: из старых фильмов в доступе Кинопоиска остались только лучшие фильмы, неудачные фильмы тех времен просто не за чем отсавлять на платформе. 

```python
print("Средняя оценка у оскароносных актеров", df[(df['OscarA']==1)]['Rate'].mean())
print("Средняя оценка у не оскароносных актеров",df[(df['OscarA']==0)]['Rate'].mean())
print("Средняя оценка у оскароносных режиссеров", df[(df['OscarD']==1)]['Rate'].mean())
print("Средняя оценка у не оскароносных режиссеров", df[(df['OscarD']==0)]['Rate'].mean())
```

Разница есть, в среднем оскароносный актер поднимает рейтинг на 0.3 балла, режиссер - на 0.6

```python
plt.scatter(df['OscarD'],df['Rate'])
plt.xlabel('Наличие оскароносного режиссера')
plt.ylabel('Рейтинг фильма')
```

Оскароносные режиссеры не снимают плохих фильмов. Да, есть не совсем удачные, с рейтингом 6-7, но низкокачесвтенный и малобюджетный фильм условный Скорсезе снимать не будет.
