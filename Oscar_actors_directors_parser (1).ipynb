```python
import requests
from requests import get
import numpy as np
import pandas as pd
import time
```

```python
page_link='https://kino.mail.ru/awards/oscar/nomination/5/'
response=get(page_link)
response
from selenium import webdriver
driver = webdriver.Chrome('./chromedriver.exe')
ref='https://kino.mail.ru/awards/oscar/nomination/5/'
driver.get(ref)
from bs4 import BeautifulSoup
soup=BeautifulSoup(driver.page_source)
```

```python
names=soup.find_all('a', {'class': 'link-holder'})
names_list=list(names)
df_names = pd.DataFrame(names_list)
df_names
```

```python
page_link2='https://kino.mail.ru/awards/oscar/nomination/5/?page=2'
response2=get(page_link2)
response2
from selenium import webdriver
driver2 = webdriver.Chrome('./chromedriver.exe')
ref2='https://kino.mail.ru/awards/oscar/nomination/5/?page=2'
driver2.get(ref2)
from bs4 import BeautifulSoup
soup2=BeautifulSoup(driver2.page_source)
```

```python
names2=soup2.find_all('a', {'class': 'link-holder'})
names_list2=list(names2)
df_names2 = pd.DataFrame(names_list2)
df_names2
```

```python
page_link3='https://kino.mail.ru/awards/oscar/nomination/5/?page=3'
response3=get(page_link3)
response3
from selenium import webdriver
driver3 = webdriver.Chrome('./chromedriver.exe')
ref3='https://kino.mail.ru/awards/oscar/nomination/5/?page=3'
driver3.get(ref3)
from bs4 import BeautifulSoup
soup3=BeautifulSoup(driver3.page_source)
```

```python
names3=soup3.find_all('a', {'class': 'link-holder'})
names_list3=list(names3)
df_names3 = pd.DataFrame(names_list3)
df_names3
```

```python
resultm = pd.concat([df_names, df_names2, df_names3], axis=0)
resultm
dfm = resultm.drop(resultm.index[1::2])
dfm
```

```python
dfm.to_csv('actresssoscar.csv')
import os
print(os.getcwd())
```
