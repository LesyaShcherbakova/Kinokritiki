```python
import requests
from requests import get
import numpy as np
import pandas as pd
import time
```

```python
page_link='https://www.kinopoisk.ru/lists/movies/popular-films/?sort=rating'
response=get(page_link)
response
from selenium import webdriver
driver = webdriver.Chrome('./chromedriver.exe')
ref='https://www.kinopoisk.ru/lists/movies/popular-films/?sort=rating'
driver.get(ref)
```

```python
from bs4 import BeautifulSoup
soup=BeautifulSoup(driver.page_source)
names=soup.find_all('span', {'class': 'styles_mainTitle__IFQyZ styles_activeMovieTittle__kJdJj'})
names_list=list(names)
df_names = pd.DataFrame(names_list)
ocenka=soup.find_all('span', {'class': 'styles_kinopoiskValuePositive__vOb2E styles_kinopoiskValue__9qXjg styles_top250Type__mPloU'})
names_info=list(ocenka)
df_ocenka = pd.DataFrame(names_info)
result1 = pd.concat([df_names, df_ocenka,], axis=1)
o=soup.find_all('span', {'class': 'desktop-list-main-info_truncatedText__IMQRP'})
names_o=list(o)
df_o = pd.DataFrame(names_o)
view=soup.find_all('span', {'class': 'styles_kinopoiskCount__2_VPQ'})
view_list=list(view)
df_view = pd.DataFrame(view)
year_duration=soup.find_all('span', {'class': 'desktop-list-main-info_secondaryText__M_aus'})
df_yd = pd.DataFrame(year_duration, columns=['movie_info'])
year=[]
duration=[]
for i in df_yd['movie_info']:
        p=str(i).replace('<span class="desktop-list-main-info_secondaryText__M_aus">','')
        p=p.replace(', <!-- -->','')
        p=p.replace('\xa0мин.</span>','')
        p=p.replace(' ','',)
        p=p.replace(',','',)
        year.append(p[:4])
        duration.append(p[4:])
df_y=pd.DataFrame(year,columns=['Year']).astype(int)
df_d=pd.DataFrame(duration,columns=['Duration']).astype(int)
result_page1 = pd.concat([result1, df_o, df_view, df_y, df_d], axis=1)
result_page1.columns=['Name','Rate','Info', 'View','Year','Duration']
result_page1
```
